```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
import sys
import os

# Proje yolunu ekle
sys.path.append(os.path.abspath(os.path.join('..')))
from src.data_loader import load_raw_data
from src.preprocessing import split_features_target, build_preprocessor

# Ayarlar
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))

# 1. Orijinal Veri
df = load_raw_data("/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/data/raw-data/Sleep_health_and_lifestyle_dataset.csv")
X, y = split_features_target(df)
original_counts = y.value_counts()

plt.subplot(1, 2, 1)
sns.barplot(x=original_counts.index, y=original_counts.values, palette="Reds")
plt.title("Orijinal Veri DaÄŸÄ±lÄ±mÄ± (Dengesiz)", fontsize=12, fontweight='bold')
plt.xlabel("SÄ±nÄ±flar")
plt.ylabel("KiÅŸi SayÄ±sÄ±")
plt.bar_label(plt.gca().containers[0]) # type: ignore

# 2. SMOTE SonrasÄ± Veri
# HÄ±zlÄ±ca SMOTE uygulayalÄ±m (GÃ¶rsel iÃ§in)
le = LabelEncoder()
y_enc = le.fit_transform(y)
preprocessor = build_preprocessor(X)
X_trans = preprocessor.fit_transform(X)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_trans, y_enc) # type: ignore
res_counts = pd.Series(y_res).value_counts() # type: ignore
res_counts.index = le.classes_ # Ä°simleri geri getir

plt.subplot(1, 2, 2)
sns.barplot(x=res_counts.index, y=res_counts.values, palette="Greens")
plt.title("SMOTE SonrasÄ± DaÄŸÄ±lÄ±m (Dengeli)", fontsize=12, fontweight='bold')
plt.xlabel("SÄ±nÄ±flar")
plt.ylabel("KiÅŸi SayÄ±sÄ±")
plt.bar_label(plt.gca().containers[0]) # type: ignore

plt.tight_layout()
plt.savefig("/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/figures/data_balance_comparison.png")
plt.show()
print("âœ… Veri daÄŸÄ±lÄ±m gÃ¶rseli kaydedildi.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os

# GÃ¶rsel AyarlarÄ±
sns.set(style="whitegrid")
plt.rcParams["figure.dpi"] = 150
plt.rcParams["font.family"] = "serif"

# Dosya YollarÄ± (Senin bilgisayarÄ±ndaki yollar)
BASE_DIR = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/tables-baseline"
FILTER_DIR = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/tables-filtered"
SMOTE_DIR = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/tables-baseline-smote"

# Dosya Ä°simleri (EÄŸer klasÃ¶r isimlerin farklÄ±ysa burayÄ± kontrol et)
FILES = {
    "Baseline": os.path.join(BASE_DIR, "baseline_metrics.csv"),
    "SMOTE": os.path.join(SMOTE_DIR, "baseline_smote_metrics.csv"),
    "Filtreli": os.path.join(FILTER_DIR, "filtered_metrics_all.csv")
}

print("ðŸ“Š Proje Evrimi Grafikleri (Model Ä°simleriyle) HazÄ±rlanÄ±yor...")

best_models = []

# 1. Her Dosyadan En Ä°yi Modeli Ã‡ek
for stage_name, file_path in FILES.items():
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        
        # En yÃ¼ksek F1-Macro skoruna sahip modeli bul
        best_row = df.sort_values(by="f1_macro", ascending=False).iloc[0]
        
        # --- MODEL VE FÄ°LTRE Ä°SMÄ°NÄ° AYIKLAMA ---
        # Model adÄ±nÄ± bul (bazÄ± dosyalarda 'model', bazÄ±larÄ±nda 'base_model' yazar)
        raw_model_name = best_row['model'] if 'model' in best_row else best_row['base_model']
        
        # Gereksiz ekleri temizle
        clean_model_name = raw_model_name.replace("_SMOTE", "").replace("_none", "").replace("_wavelet", "")
        
        # Filtre var mÄ± kontrol et (Sadece filtered dosyasÄ±nda vardÄ±r)
        filter_suffix = ""
        if "filter_method" in best_row:
            filter_method = str(best_row["filter_method"])
            if filter_method.lower() != "none":
                filter_suffix = f" + {filter_method.capitalize()}" # Ã–rn: + Wavelet
            else:
                filter_suffix = " (Raw)" # Filtre yoksa Raw yazalÄ±m
        
        # SonuÃ§ Ä°sim: "SVM" veya "RF + Wavelet"
        display_name = f"{clean_model_name}{filter_suffix}"
        
        # Grafikte gÃ¶rÃ¼necek tam etiket: "Filtreli [RF + Wavelet]"
        full_label = f"{stage_name}\n[{display_name}]"
        
        # Verileri topla
        metrics = {
            "AÅŸama": full_label, # Hue (Renk) olarak bunu kullanacaÄŸÄ±z
            "Model": clean_model_name,
            "Accuracy": best_row.get("accuracy", 0),
            "F1 Score (Macro)": best_row.get("f1_macro", 0),
            "Precision (Macro)": best_row.get("precision_macro", 0),
            "Recall (Macro)": best_row.get("recall_macro", 0),
            "Specificity": best_row.get("specificity_macro", 0),
            "AUC": best_row.get("roc_auc_ovr_macro", 0)
        }
        best_models.append(metrics)
    else:
        print(f"âš ï¸ UyarÄ±: {file_path} bulunamadÄ±. LÃ¼tfen dosya yolunu kontrol edin.")

if best_models:
    df_evolution = pd.DataFrame(best_models)
    
    # Veriyi "Long Format"a Ã§evir
    df_melted = df_evolution.melt(
        id_vars=["AÅŸama", "Model"], 
        var_name="Metrik", 
        value_name="Skor"
    )

    # --- GÃ–RSEL 1: GRUPLU BAR GRAFÄ°ÄžÄ° ---
    plt.figure(figsize=(14, 7))
    
    chart = sns.barplot(
        data=df_melted,
        x="Metrik",
        y="Skor",
        hue="AÅŸama", # ArtÄ±k burada model ismi de yazacak
        palette=["#95a5a6", "#3498db", "#2ecc71"] # Gri -> Mavi -> YeÅŸil
    )
    
    plt.title("Proje GeliÅŸim SÃ¼reci: Åžampiyon Modeller ve PerformanslarÄ±", fontsize=16, fontweight='bold')
    plt.ylim(0.5, 1.02) 
    plt.legend(title="AÅŸama [Kazanan Model]", loc="lower right", frameon=True)
    plt.ylabel("Skor (0-1 ArasÄ±)")
    plt.xlabel("")
    
    for container in chart.containers:
        chart.bar_label(container, fmt="%.3f", padding=3, fontsize=9, fontweight='bold')

    plt.tight_layout()
    save_path_1 = "../results/figures/evolution_all_metrics_named.png"
    plt.savefig(save_path_1)
    plt.show()
    print(f"âœ… Evrim grafiÄŸi kaydedildi: {save_path_1}")
    
    # --- GÃ–RSEL 2: DETAYLI HEATMAP ---
    plt.figure(figsize=(12, 5))
    
    # Pivot tablo yap (Index artÄ±k model ismini de iÃ§eriyor)
    heatmap_data = df_evolution.set_index("AÅŸama").drop(columns=["Model"])
    
    sns.heatmap(
        heatmap_data,
        annot=True,
        fmt=".3f",
        cmap="Blues",
        linewidths=1,
        cbar=False,
        annot_kws={"size": 13, "weight": "bold"}
    )
    
    plt.title("Åžampiyon Modellerin DetaylÄ± Skor KarÅŸÄ±laÅŸtÄ±rmasÄ±", fontsize=15, fontweight='bold')
    plt.yticks(rotation=0) # SatÄ±r isimlerini dÃ¼z yap
    plt.tight_layout()
    
    save_path_2 = "../results/figures/evolution_heatmap_named.png"
    plt.savefig(save_path_2)
    plt.show()
    print(f"âœ… Detay tablosu kaydedildi: {save_path_2}")

    print("\n--- KAZANANLAR LÄ°STESÄ° ---")
    print(df_evolution[["AÅŸama", "F1 Score (Macro)"]])

else:
    print("âŒ HiÃ§bir veri dosyasÄ± okunamadÄ±. YollarÄ± kontrol et!")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import math

# GÃ¶rsel AyarlarÄ±
sns.set(style="white")
plt.rcParams["figure.dpi"] = 150
plt.rcParams["font.family"] = "serif"

# --- DOSYA YOLLARI ---
# SMOTE sonuÃ§larÄ±nÄ±n olduÄŸu klasÃ¶r (run_baseline_smote.py buraya kaydetmiÅŸti)
# EÄŸer klasÃ¶r ismini deÄŸiÅŸtirdiysen burayÄ± gÃ¼ncelle!
METRICS_PATH = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/tables-baseline-smote/baseline_smote_metrics.csv"
TABLES_DIR = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/tables-baseline-smote" 

print("ðŸ“Š SMOTE Modellerinin Hata Matrisleri Ã‡iziliyor...")

if os.path.exists(METRICS_PATH):
    # 1. Metrikleri Oku
    df_metrics = pd.read_csv(METRICS_PATH)
    # Accuracy skoruna gÃ¶re sÄ±ralayalÄ±m ki en iyiler baÅŸta Ã§Ä±ksÄ±n.
    df_sorted = df_metrics.sort_values(by = "accuracy", ascending = False)
    models = df_sorted["model"].tolist()

    print(f"Toplam {len(models)} model bulundu: {models}")



    # 2. Grid AyarlarÄ±
    num_plots = len(models)
    n_cols = 3
    n_rows = math.ceil(num_plots / n_cols)
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten() 
    
    cmap = "Blues" # SMOTE iÃ§in Mavi tonu (GÃ¼veni/Ä°stikrarÄ± simgelesin)

    for i, model_name in enumerate(models):
        score = df_sorted.loc[df_sorted['model'] == model_name, 'f1_macro'].values[0]
        
        # Dosya adÄ±nÄ± oluÅŸtur: confusion_matrix_SVM_SMOTE.csv
        cm_filename = f"confusion_matrix_{model_name}.csv"
        cm_path = os.path.join(TABLES_DIR, cm_filename)
        
        ax = axes[i]
        
        if os.path.exists(cm_path):
            cm_df = pd.read_csv(cm_path, index_col=0)
            
            sns.heatmap(
                cm_df, 
                annot=True, 
                fmt="d", 
                cmap=cmap, 
                cbar=False, 
                ax=ax, 
                linewidths=1, 
                linecolor='lightgray',
                annot_kws={"size": 14, "weight": "bold"}
            )
            
            # BaÅŸlÄ±ÄŸÄ± temizle (SVM_SMOTE -> SVM + SMOTE)
            clean_name = model_name.replace("_SMOTE", " + SMOTE")
            ax.set_title(f"{clean_name}\n(F1: {score:.3f})", fontweight='bold', size=12, pad=10)
            ax.set_xlabel(""); ax.set_ylabel("")
            
        else:
            ax.text(0.5, 0.5, "Dosya Yok", ha='center', va='center', color='red')
            ax.set_title(model_name)

    # BoÅŸ kalan grafikleri temizle
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # Ortak Eksen Ä°simleri
    fig.text(0.5, 0.02, 'Tahmin Edilen (Predicted)', ha='center', fontsize=14, fontweight='bold')
    fig.text(0.02, 0.5, 'GerÃ§ek Etiket (True)', va='center', rotation='vertical', fontsize=14, fontweight='bold')

    plt.subplots_adjust(hspace=0.4, wspace=0.3)
    plt.tight_layout(rect=[0.05, 0.05, 1, 1])
    
    save_path = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/figures/confusion_matrices_smote.png"
    plt.savefig(save_path, bbox_inches='tight')
    plt.show()
    
    print(f"âœ… SMOTE gÃ¶rselleri kaydedildi: {save_path}")

else:
    print(f"âŒ Dosya bulunamadÄ±: {METRICS_PATH}")
```
