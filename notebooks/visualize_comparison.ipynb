```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# GÃ¶rsel AyarlarÄ±
sns.set(style="whitegrid")
plt.rcParams["figure.dpi"] = 150
plt.rcParams["font.family"] = "serif"

# Dosya YollarÄ±
BASELINE_PATH = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/tables-baseline/baseline_metrics.csv"
SMOTE_PATH = "/Users/melih/Desktop/sleep-project/sleep-disorder-signal-processing/results/tables-baseline-smote/baseline_smote_metrics.csv"

print("ðŸ“Š Baseline vs SMOTE KarÅŸÄ±laÅŸtÄ±rmasÄ± HazÄ±rlanÄ±yor...")

if os.path.exists(BASELINE_PATH) and os.path.exists(SMOTE_PATH):
    # 1. Verileri Oku
    df_base = pd.read_csv(BASELINE_PATH)
    df_smote = pd.read_csv(SMOTE_PATH)
    
    # 2. Etiketleme (Hangi deney olduÄŸunu belirtelim)
    df_base["Experiment"] = "Baseline (Ham Veri)"
    df_smote["Experiment"] = "SMOTE (DengelenmiÅŸ)"
    
    # 3. Ä°sim TemizliÄŸi
    # SMOTE tablosunda isimler "SVM_SMOTE" gibi, onlarÄ± "SVM" yapalÄ±m ki eÅŸleÅŸsin
    df_smote["model"] = df_smote["model"].str.replace("_SMOTE", "")
    
    # 4. TablolarÄ± BirleÅŸtir (Alt alta ekle)
    df_compare = pd.concat([df_base, df_smote], ignore_index=True)
    
    # 5. SÄ±ralama (Baseline F1 skoruna gÃ¶re sÄ±ralayalÄ±m)
    # Ã–nce modelleri baseline skoruna gÃ¶re sÄ±ralÄ± bir liste yapalÄ±m
    order = df_base.sort_values(by="f1_macro", ascending=False)["model"].tolist()
    
    # 6. Ã‡izim
    plt.figure(figsize=(14, 7))
    
    # Bar Plot (hue='Experiment' sayesinde yan yana Ã§izer)
    ax = sns.barplot(
        data=df_compare,
        x="model",
        y="f1_macro",
        hue="Experiment",
        order=order, # SÄ±ralamayÄ± uygula
        palette=["#34495e", "#2ecc71"] # Koyu Mavi (Baseline) vs YeÅŸil (SMOTE)
    )
    
    # BaÅŸlÄ±k ve Eksenler
    plt.title("Veri Dengeleme Etkisi: Baseline vs SMOTE (F1-Score Macro)", fontsize=16, fontweight='bold')
    plt.xlabel("Model", fontsize=12)
    plt.ylabel("F1 Score (Macro)", fontsize=12)
    plt.ylim(0, 1.15) # Ãœstte boÅŸluk bÄ±rak
    plt.legend(title="Deney Grubu", loc="upper right")
    
    # 7. DeÄŸerleri YazdÄ±rma
    for container in ax.containers:
        ax.bar_label(container, fmt="%.2f", padding=3, fontsize=9, fontweight='bold')

    plt.tight_layout()
    
    # Kaydet
    save_path = "../results/figures/comparison_baseline_vs_smote_f1_macro.png"
    plt.savefig(save_path, bbox_inches='tight')
    plt.show()
    
    print(f"âœ… KarÅŸÄ±laÅŸtÄ±rma grafiÄŸi kaydedildi: {save_path}")

else:
    print("âŒ Dosyalar bulunamadÄ±! LÃ¼tfen Ã¶nce run_baseline.py ve run_baseline_smote.py Ã§alÄ±ÅŸtÄ±rÄ±n.")
```
